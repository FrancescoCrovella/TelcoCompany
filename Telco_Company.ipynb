```python
def countplotting(x,data):
    plt.figure(figsize=(18,5))
    chart=sns.countplot(x=x,data=data)
    
    total=float(len(data))
    
    for value in chart.patches:
        xi=value.get_x()+value.get_width()
        yi=value.get_height()
        percentage_value='{:.2f}%'.format(100*yi/total)
        chart.annotate(percentage_value,(xi,yi))
```
